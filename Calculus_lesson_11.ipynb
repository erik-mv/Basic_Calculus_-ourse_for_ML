## ДУ с разделяющимися переменными

### $$N(x)\,\mbox{d}x+M(y)\,\mbox{d}y=0$$

Это самый простой вид ДУ. Из названия понятно, что переменные в таком уравнении можно разделить

Пример:

1. 
### $$xy\,\mbox{d}x+(x+1)\,\mbox{d}y=0$$

### $$\frac{x}{x+1}\,\mbox{d}x+\frac{1}{y}\,\mbox{d}y=0,\,\,\,\, \mbox{предполагая,что}\,\,  x+1\neq0 \,\,\mbox{и}\,\, y\neq0$$

### $$\int\frac{x}{x+1}\,\mbox{d}x+\int\frac{1}{y}\,\mbox{d}y=0$$

### $$\int\frac{x+1-1}{x+1}\,\mbox{d}x+\int\frac{1}{y}\,\mbox{d}y=0$$

### $$\int \Bigl(1-\frac{1}{x+1}\Bigl)\,\mbox{d}x+\int\frac{1}{y}\,\mbox{d}y=0$$

### $$x-\ln|x+1|+\ln|y|=C$$

## Задача Коши

Из предыдущего примера можно сделать выводы, что __решением ДУ является семейство функций__.

Есть задачи с условием прохождения через точку, то есть семейства функций, выбирается одна. Такие задачи, называются задачами Коши.

2. 
### $$(x^2-1)\,y'+2xy^2=0,\,\,\,\,\,\, y(0)=1$$

### $$(x^2-1)\frac{\mbox{d}y}{\mbox{d}x}+2xy^2=0$$

### $$\frac{1}{y^2}\,\mbox{d}y+\frac{2x}{x^2-1}\,\mbox{d}x=0,\,\,\,\, \mbox{предполагая,что}\,\,  y^2\neq0 \,\,\mbox{и}\,\, x^2-1\neq0$$

### $$\int\frac{1}{y^2}\,\mbox{d}y+\int\frac{2x}{x^2-1}\,\mbox{d}x=0$$


### $$-\frac{1}{y}+\ln|x^2-1|=C$$

Воспользуемся условием $y(0)=1$

### $$-\frac{1}{1}+\ln|0^2-1|=C \,\,\,\,\,\Rightarrow\,\,\,\,\, C=-1$$

Тогда решением будет функция:

### $$-\frac{1}{y}+\ln|x^2-1|=-1$$

### $$\frac{1}{y}=\ln|x^2-1|+1$$

### Однородные ДУ

Однородные - это когда все члены уравнения в одной степени

### $$(x + 2y)\, \mbox{d}x − x\,\mbox{d}y = 0$$

### $$(y^2 − 2xy) \,\mbox{d}x + x^2 \,\mbox{d}y = 0$$

Решаются такие уравнения через замену 

### $$y(x)=x\cdot t(x) \,\,\,\,\,\Rightarrow\,\,\,\,\, \mbox{d}y=x\,\mbox{d}t+t\,\mbox{d}x$$

После замен получается ДУ с раздающимися переменными

3. 
### $$(x + 2y)\, \mbox{d}x − x\,\mbox{d}y = 0$$

### $$(x + 2xt)\, \mbox{d}x − x(x\,\mbox{d}t+t\,\mbox{d}x) = 0$$

### $$(x + xt)\, \mbox{d}x − x^2\,\mbox{d}t= 0$$

В данном уравнении на $x$делить нельзя, т.к. он сократится, поэтому вынесем его за скобку и получим одно из решений.

### $$x\Bigl((1 + t)\, \mbox{d}x − x\,\mbox{d}t\Bigr)= 0$$

### $$x=0\,\,\, \mbox{или} \,\,\,(1 + t)\, \mbox{d}x − x\,\mbox{d}t= 0$$

### $$\frac{1}{x}\, \mbox{d}x − \frac{1}{t+1}\,\mbox{d}t= 0$$

### $$\int\frac{1}{x}\, \mbox{d}x − \int\frac{1}{t+1}\,\mbox{d}t= 0$$


### $$\ln|x| − \ln|t+1|= \ln C$$

### $$\ln|x| − \ln\Bigl|\frac{y}{x}+1\Bigr|= \ln C$$

### $$\ln|x| − \ln\Bigl|\frac{y+x}{x}\Bigr|= \ln C$$

### $$\ln\Bigl|\frac{x^2}{y+x}\Bigr|= \ln C$$

### $$\frac{x^2}{y+x}=C$$

### $$y+x=\frac{x^2}{C}$$

### $$y=\frac{x^2}{C}-x\,\,\, \mbox{или} \,\,\,x=0$$

### линейные ДУ первого порядка;

Если дифференциальное уравнение можно представить в следующем виде, то оно является линейным

### $$y'+a(x)y=b(x)$$

Линейные уравнения решатся через однородные, с занулением правой части. При этом уравнение получается даже не однородное, а сразу с разделяющимися переменными.

4. 
### $$x\cdot y'+y=2x$$

### $$y'+\frac{y}{x}=2,\,\,\,\,\,\,x\neq0$$

### Метод вариации произвольных постоянных

### $$y'+\frac{y}{x}=0$$

### $$\frac{\mbox{d}y}{\mbox{d}x}+\frac{y}{x}=0$$

### $$\frac{\mbox{d}y}{y}+\frac{\mbox{d}x}{x}=0$$

### $$\int\frac{\mbox{d}y}{y}+\int\frac{\mbox{d}x}{x}=0$$


### $$\ln|y|+\ln|x|=\ln C$$

### $$yx=C$$

### $$y=\frac{C}{x}$$

Данное решение называется общим. Т.е. общее для всех левых частей уравнения $y'+\frac{y}{x}=b(x)$

Но мы допустили ошибку занулив правую часть, ошибка кроется в константе, ведь на самом деле это может быть функция, зависимая от $x$.

### $$y=\frac{C(x)}{x}$$

Чтобы найти $C(x)$ подставим общее решение в наше уравнения.

### $$y'+\frac{y}{x}=2$$

### $$\frac{C'(x)\cdot x-C(x)}{x^2}+\frac{C(x)}{x^2}=2$$

### $$\frac{C'(x)\cdot x}{x^2}-\frac{C(x)}{x^2}+\frac{C(x)}{x^2}=2$$


### $$\frac{C'(x)\cdot x}{x^2}=2$$

### $$C'(x)=2x$$

### $$\frac{\mbox{d}C}{\mbox{d}x}=2x$$

### $$\mbox{d}C=2x\,dx$$

### $$C(x)=x^2+C_2$$

Подставим найденную константу в общее решение:

### $$y=\frac{x^2+C_2}{x}$$

### $$y=x+\frac{C_2}{x}$$

Полученное решение называется частным.

### Уравнение в полных дифференциалах

### $$\mbox{d}\Bigl(U(x,y)\Bigr)=U'_x\,\mbox{d}x+U'_y\,\mbox{d}y$$

### $$U'_x\,\mbox{d}x+U'_y\,\mbox{d}y=0$$

Для проверки достаточно найти смешанные производные и проверить их равенство

5. 
### $$\Bigl(4x^2+\frac{1}{2}y^2-6y\Bigr)\,\mbox{d}x+\Bigl(2y^3+xy-6x\Bigr)\,\mbox{d}y=0$$

### $$U'_x=4x^2+\frac{1}{2}y^2-6y\,\,\,\,\,\Rightarrow\,\,\,\,\, U''_{xy} = y-6$$

### $$U'_y=2y^3+xy-6x\,\,\,\,\,\Rightarrow\,\,\,\,\, U''_{yx} = y-6$$

Так как смешанные производные равны, то перед нами действительно уравнение в полных дифференциалах

Решается оно довольно просто, с помощью одного из условий мы восстанавливаем функцию, с помощью другого находим недостающие элементы:

### $$U'_x=4x^2+\frac{1}{2}y^2-6y\,\,\,\,\,\Rightarrow\,\,\,\,\, U=\int \Bigl(4x^2+\frac{1}{2}y^2-6y\Bigr)\,dx$$

### $$U=\frac{4}{3}x^3+\frac{1}{2}y^2x-6yx+\varphi(y)$$

Чтобы найти $\varphi(y)$ воспользуемся вторым условием

### $$U'_y=yx-6x+\varphi'(y)=2y^3+xy-6x$$

### $$\varphi'(y)=2y^3$$

### $$\varphi(y)=\frac{y^4}{2}+C$$

Получаем решение 

### $$\frac{4}{3}x^3+\frac{1}{2}y^2x-6yx+\frac{y^4}{2}=C$$
