__9**.__ Написать на python алгоритм, по вычислению значений $\sin(x)$ для $х \in[0, 30^o]$

Для проверки подойдут данные из таблицы Брадиса

$$\sin(0)=0.0000$$

$$\sin(1^0)=0.0175$$

$$\sin(2^0)=0.0349$$

$$\sin(3^0)=0.0523$$

$$\sin(4^0)=0.0698$$
	
$$\sin(5^0)=0.0872$$

Решение:

```python
import numpy as np
from scipy.interpolate import splev, splrep

n = 10000
l = np.pi / 6
h = l / n
x = np.arange(0, l, h)
x = np.append(x,l)
sin_d = np.array([0])
cos_d = 1
    
while(n != 0):
    sin_d = np.append(sin_d, sin_d[-1] + cos_d * h)
    cos_d = np.sqrt(1 - (sin_d[-1])**2)
    n -= 1
    
spl = splrep(x, sin_d)

def sin(k):
    return splev(k, spl)
```

```python
alfa = 30

print(sin(alfa * np.pi / 180))
print(np.sin(alfa * np.pi / 180))
```
