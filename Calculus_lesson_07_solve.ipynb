### $$\begin{pmatrix}

теперь можно вычислить главные миноры 

### $$\Delta_{1}=U''_{xx}=2y+4$$

### $$\Delta_{2}=\begin{vmatrix}
U''_{xx} & U''_{xy}\\ 
U''_{yx} & U''_{yy}\\
\end{vmatrix}=U''_{xx}\cdot U''_{yy}-(U''_{xy})^2=4y^2+20y+24-4x^2$$

Осталось проверить знаки миноров для каждой точки.

### $$\Delta_{1}=2y+4=4>0$$

Как правило дальнейшее иследование через матрицу Гёссе не имеет смысла. Но давайте пересобирем матрицу Гёссе

### $$\begin{pmatrix}
U''_{yy} & U''_{yx}\\ 
U''_{xy} & U''_{xx}\\
\end{pmatrix}$$



### $$\begin{pmatrix}
2y+6 & 2x\\ 
2x & 2y+4\\
\end{pmatrix}$$

Тогда 

### $$\Delta_{1}=U''_{yy}=2y+6=2\cdot(-2)+6=2>0$$

### $$\Delta_{2}=\begin{vmatrix}
U''_{yy} & U''_{yx}\\

\end{vmatrix}=U''_{xx}\cdot U''_{yy}-(U''_{xy})^2=4y^2+20y+24-4x^2=$$

### $$=4\cdot4+20\cdot(-2)+24-4\cdot8=40-40-32<0$$

__7*.__ Исследовать на экстремум функцию:

### $$U=e^{-\frac{x}{2}}\cdot(x^2+y^2)$$

Решение:

Найдем частные производные, приравняем их к нулю и найдем стационарные точки:

### $$\begin{cases}
   U'_{x}=-\frac{1}{2}e^{-\frac{x}{2}}(x^2+y^2)+e^{-\frac{x}{2}}\cdot2x=0 \\
   U'_{y}=e^{-\frac{x}{2}}\cdot2y=0 
 \end{cases}\,\,\,\,\,\Rightarrow$$
 
### $$\Rightarrow\,\,\,\,\,\begin{cases}
   -\frac{1}{2}(x^2+y^2)+2x=0 \\
   y=0 
 \end{cases}\,\,\,\,\,\Rightarrow\,\,\,\,\,
 \begin{cases}
   x^2-4x=0 \\
   y=0 
 \end{cases}\,\,\,\,\,\Rightarrow\,\,\,\,\,
 \begin{gathered} 
(0,0)\\
(4,0)
\end{gathered}$$
 
Найдем вторые производные и составим матрицу Гёссе в общем виде:
 
### $$U''_{xx}=\frac{1}{4}e^{-\frac{x}{2}}(x^2+y^2)-\frac{1}{2}e^{-\frac{x}{2}}\cdot2x-\frac{1}{2}e^{-\frac{x}{2}}\cdot2x+2e^{-\frac{x}{2}}$$ 
### $$U''_{xy}=U''_{yx}=-\frac{1}{2}e^{-\frac{x}{2}}\cdot2y$$  
### $$U''_{yy}=2e^{-\frac{x}{2}}$$

Осталось проверить знаки миноров для каждой точки. 

Для точки $(0,0)$

### $$\Delta_{1}=2>0$$

### $$\Delta_{2}=
\begin{vmatrix}
2 & 0\\ 
0 & 2\\
\end{vmatrix}=4>0$$

Т.е. $(0,0)$ - точка минимума

Для точки $(4,0)$

### $$\Delta_{1}=-\frac{2}{e^2}<0$$

### $$\Delta_{2}=
\begin{vmatrix}
-\frac{2}{e^2} & 0\\ 
0 & \frac{2}{e^2}\\
\end{vmatrix}=-\frac{4}{e^2}<0$$

Т.е. $(4,0)$ - седловая точка

```python
def makeData ():
    x = np.arange (-8, 8, 0.1)
    y = np.arange (-8, 8, 0.1)
    xgrid, ygrid = np.meshgrid(x, y)

    Ugrid = np.e**(-xgrid / 2) * (xgrid**2 + ygrid**2)
    return xgrid, ygrid, Ugrid

x, y, U = makeData()

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(projection='3d')
ax.plot_surface(x, y, U)

plt.show()
```

__8**.__ С помощью метода наименьших квадратов (МНК) подобрать значения параметров $a$ и $b$ для сигмоидальной функции 

### $$y=\frac{1}{1+e^{-(ax+b)}}$$ 

по заданному распределению. Построить график.

```python
# Задаем рандомное распределение точек
xi = np.random.uniform(-8, 8, 200)
yi = 1 / (1 + np.exp(-np.random.uniform(0.5, 1.5, 200) * xi - np.random.uniform(-1, 1, 200)))

plt.figure(figsize=(16, 8))
plt.plot(xi,yi, marker="o", ls="")
plt.show()
```

Решение

```python
# Задаем рандомное распределение точек
xi = np.random.uniform(-8, 8, 200)
yi = 1 / (1 + np.exp(-np.random.uniform(0.5, 1.5, 200) * xi - np.random.uniform(-1, 1, 200)))

plt.figure(figsize=(16, 8))
plt.plot(xi,yi, marker="o", ls="")


def equations(p):
    a, b = p
    # Это заранее найденные производные от (y-f(x))**2 по переменным a и b
    exp = np.exp(-a * xi - b)
    return (
        (-2 * xi * exp * (yi - 1 / (1 + exp)) / (1 + exp)**2).sum(),
        (-2 * exp * (yi - 1 / (1 + exp)) / (1 + exp)**2).sum(),
    )
# Численное решение нелинечной системы уравнений
a, b =  fsolve(equations, (0, 0))
print (a, b)

x = np.linspace(-8, 8, 200)
plt.plot(x, 1/(1+np.exp(-a * x - b)))
plt.show()
```
